# Introduction

## Apa itu Tidur?
Tidur adalah periode istirahat.

Tidur bukan hanya saat ketika otak dan tubuh Anda mati. Mendapatkan cukup tidur membantu Anda berpikir lebih jernih dan bereaksi lebih cepat. Tidak mendapatkan cukup tidur bisa berbahaya, tidak hanya memengaruhi kinerja Anda, tetapi juga kesehatan dan mood Anda.

## Pentingnya Tidur
Ketika kamu tertidur, otak tetap sibuk untuk mengawasi kerja sistem-sistem di dalam tubuh untuk menghasilkan energi dan meningkatkan kesehatan secara keseluruhan. Jumlah waktu tidur yang dibutuhkan setiap orang sebenarnya tergantung pada sejumlah faktor, terutama usia.

Tidur yang cukup memungkinkan sistem kekebalan tubuh berfungsi optimal dengan memproduksi sitokin, yaitu senyawa yang membantu melawan bakteri dan virus dalam tubuh.

Beberapa orang mengaku dapat cukup beristirahat hanya dengan beberapa jam tidur semalam saja. Padahal, hal ini berbahaya dan memengaruhi kinerja dan aktivitas sehari-hari. Orang-orang yang tidur sangat sedikit di malam hari cenderung kesulitan untuk melakukan tugas yang rumit. Oleh sebab itu, pastikan kamu mendapatkan waktu tidur yang cukup untuk menghindari kondisi yang tidak diinginkan.

## Tujuan Proyek
Proyek ini bertujuan untuk mengungkap penyebab insomnia dan sleep apnea, memberikan wawasan yang lebih jelas tentang apa yang menyebabkan gangguan tidur ini.

Mengingat pentingnya tidur untuk kesehatan fisik dan mental, proyek ini juga bertujuan untuk mengembangkan strategi yang dapat membantu individu mengidentifikasi dan memodifikasi faktor-faktor risiko dalam kehidupan mereka yang dapat memperburuk gangguan tidur. 

Dengan demikian, proyek ini tidak hanya berfokus pada penyebab masalah tidur, tetapi juga pada solusi praktis yang dapat membantu meningkatkan kualitas tidur dan, pada akhirnya, kesejahteraan umum. 

Tujuan akhirnya adalah untuk memberikan bantuan yang sangat dibutuhkan bagi mereka yang berjuang dengan gangguan tidur, melalui penelitian yang terinformasi dan berbasis data.

# Data Overview 
* **Person ID**: Pengenal untuk setiap individu.
* **Gender**: Jenis kelamin individu (Laki-laki/Perempuan).
* **Age**: Usia individu dalam tahun.
* **Occupation**: Pekerjaan atau profesi individu.
* **Sleep Duration (hours)**: Jumlah jam tidur individu per hari.
* **Quality of Sleep (scale: 1-10)**: Penilaian subjektif tentang kualitas tidur, rentang nilai 1 hingga 10.
* **Physical Activity Level (minutes/day)**: Jumlah menit individu melakukan aktivitas fisik setiap hari.
* **Stress Level (scale: 1-10)**: Penilaian subjektif tentang tingkat stres yang dialami individu, rentang nilai 1 hingga 10.
* **BMI Category**: Kategori IMT individu (misalnya: Kurus, Normal, Kelebihan Berat Badan).
* **Blood Pressure (systolic/diastolic)**: Pengukuran tekanan darah individu, diindikasikan sebagai tekanan sistolik atas tekanan diastolik.
* **Heart Rate (bpm)**: Denyut jantung istirahat individu dalam denyut per menit.
* **Daily Steps**: Jumlah langkah yang diambil individu per hari.
* **Sleep Disorder**: Kehadiran atau ketiadaan gangguan tidur pada individu (Tidak Ada, Insomnia, Sleep Apnea).

**Rincian Kolom Gangguan Tidur:**
* **None**: Individu tidak menunjukkan gangguan tidur spesifik.
* **Insomnia**: Individu mengalami kesulitan tidur atau tetap tidur, mengakibatkan tidur yang tidak memadai atau berkualitas buruk.
* **Sleep Apnea**: Individu mengalami hentian napas selama tidur, mengakibatkan pola tidur terganggu dan risiko kesehatan potensial.

# Data Processing

```python
import numpy as np 
import pandas as pd 
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Sleep_health_and_lifestyle_dataset.csv')
df.sample(10)
```

```python
# Daftar nama kolom di DataFrame
df.columns
```

```python
# Ringkasan struktur data di DataFrame
df.info()
```

```python
# Mengisi nilai-nilai kosong (NaN) dalam kolom 'Sleep Disorder' dengan 'None'# 
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
df.sample(10)
```

```python
# Ringkasan statistik dari kolom-kolom numerik dalam DataFrame.
df.describe()
```

# Exploratory Data Analysis

```python
# Style
color = sns.color_palette("tab10", 3)
grad = sns.color_palette("rocket", as_cmap=True)
```

**Sleep Disorder Distribution**

---

```python
sleep_count = df['Sleep Disorder'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sleep_count, autopct='%1.1f%%', startangle=135)
plt.title("Sleep Disorder Distribution", loc='left')
plt.legend(labels=sleep_count.index, title="Sleep Disorder Type", loc="upper right")
plt.show()
```

* Lebih banyak orang yang tidak memiliki gangguan tidur.

**Sleep Disorder by on Gender**

---

```python
sleep_gender = df.groupby('Sleep Disorder')['Gender'].value_counts()
sleep_gender = sleep_gender.reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x = 'Gender',
            y = 'count',
            hue = 'Sleep Disorder',
            data = sleep_gender,
            palette = color)
sns.set_style('ticks')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Sleep Disorder by Gender', loc='left')
plt.legend(title='Sleep Disorder Type')

plt.show()
```

* Tampaknya lebih banyak Pria yang tidak memiliki (**None**) gangguan tidur.
* **Insomnia** terjadi secara seimbang di antara laki-laki dan perempuan.
* Perempuan cenderung memiliki lebih banyak **Sleep Apnea** daripada Pria.

**Sleep Disorder by on Occupation**

---

```python
sleep_occupation = df.groupby('Sleep Disorder')['Occupation'].value_counts()
sleep_occupation = sleep_occupation.reset_index()
sleep_occupationt = sleep_occupation.sort_values(by='Occupation', ascending=True)

plt.figure(figsize=(8, 6))
sns.barplot(x = 'count',
            y = 'Occupation',
            hue = 'Sleep Disorder',
            data = sleep_occupation,
            palette = color)

plt.xlabel('Count')
plt.ylabel('Occupation')
plt.title('Sleep Disorder by Occupation', loc='left')
plt.legend(title='Sleep Disorder Type')

plt.show()
```

* **Promotor** ternyata paling banyak mengalami **Insomnia**, diikuti oleh **Guru** dan **Akuntan**.
* **Suster** memiliki **Sleep Apnea** lebih banyak dibandingkan dengan profesi lainnya.
* **Dokter**, **Insinyur**, **Pengacara** tidak memiliki gangguan tidur.

**Age Distribution by Sleep Disorder**

---

```python
sleep_list = ['None', 'Sleep Apnea', 'Insomnia']

fig, axes = plt.subplots(1, len(sleep_list), figsize=(16, 6), sharey=True)

for i, _ in enumerate(sleep_list):
    ax = axes[i]
    colors = color[i % len(color)]
    sns.histplot(data = df[df['Sleep Disorder'] == _], 
                 x = 'Age', 
                 kde = True, 
                 color = colors, 
                 ax = ax)
    ax.set_title(_)

plt.tight_layout()
plt.show()
```

* Pada rentang usia **30 hingga 42 tahun**, mayoritas tidak memiliki gangguan tidur (**None**).
* **Sleep Apnea** lebih umum terjadi pada rentang usia **50 hingga 57 tahun**.
* Rentang usia **43 hingga 45 tahun** memiliki kecenderungan lebih tinggi mengalami **Insomnia**.

```python
# converting non-numeric data (String or Boolean) into numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

col_corr = ['Age', 'Gender', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 
            'Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate', 
            'Daily Steps', 'Sleep Disorder']

df_cor = df.copy()

for x in col_corr:
    df_cor[x]=le.fit_transform(df_cor[x])

correlation_matrix = df_cor[col_corr].corr()
correlation_matrix
```

```python
sns.heatmap(data = correlation_matrix,
            cmap = grad,
            annot = True, 
            fmt = ".2f",
            annot_kws = {'size': 6},
            center = 0)
plt.show()
```
